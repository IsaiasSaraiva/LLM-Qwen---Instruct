```python
import torch, os, getpass
from transformers import pipeline, AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
```

```python
os.environ["HF_TOKEN"] = getpass.getpass()
```

```python
id_model = "Qwen/Qwen2.5-7B-Instruct"
```

```python
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype="float16",
    bnb_4bit_use_double_quant=True,
)
```

```python
model = AutoModelForCausalLM.from_pretrained(
    id_model,
    quantization_config= bnb_config,
    device_map="cuda",
    torch_dtype="auto",
    trust_remote_code = True,
    attn_implementation="eager"
)
```

```python
tokenizer = AutoTokenizer.from_pretrained(id_model)
```

```python
pipe = pipeline("text-generation", model = model, tokenizer = tokenizer)
```

```python
generation_args = {
    "max_new_tokens": 500,
    "return_full_text": False,
    "temperature": 0.7, # 0.1 até 0.9
    "do_sample": True,
}
```

```python
prompt = "O que é computação quântica"
#prompt = "Quanto é 7 x 6 - 42?"

output = pipe(prompt, **generation_args)
```

```python
print(output[0]['generated_text'])
```

```python
prompt = "Gere um código de arduino para leitura de Sensor de Temperatura DS18B20 a Prova D'água?"

template = """<|system|>
You are a helpful assistant.<|end|>
<|user|>
"{}"<|end|>
<|assistant|>""".format(prompt)

output = pipe(template, **generation_args)
print(output[0]['generated_text'])
```
